```python
import pandas as pd
import matplotlib.pyplot as plt
import requests
import io
```

```python
url = "https://raw.githubusercontent.com/angkohtenko/Waste_Canada/main/Source%20Data/38100138.csv"
file = requests.get(url).content
data = pd.read_csv(io.StringIO(file.decode('utf-8')))
data.head()
```

```python
data.info()
```

```python
# some materials miss sources
data[data.VALUE.isna()]
```

```python
# coordinate field represents hierarchy: .1 - all sources, .2 - residential, .3 - non-residential
data.sort_values(["REF_DATE", "COORDINATE"], inplace=True)
data.reset_index(inplace=True, drop=True)
data
```

Some materials don't have a source due to suppressed values or untracked sources.

Let's add an "Unknown source" row to the data if the source of materials diverted is not defined in the residential or non-residential rows.
In that case, the value for "Unknown source" is equal to "All sources of diverted materials".

```python
# select material rows with total amount that aren't null
unknown_source = data.loc[(data["Sources of materials prepared for recycling"]=="All sources of diverted materials") & (data['VALUE'].notna())]

# drop 'All materials diverted' row, as total split between sources will be calculated later
unknown_source = unknown_source.loc[unknown_source["Type of materials diverted"]!="All materials diverted"]

# replace Source value to 'Unknown source' if source is not described in residenatial or non-residential rows
# as df were sorted by coordinate, now first row is all sources, second row is residential and third row is non-residential
unknown_source["Sources of materials prepared for recycling"] = ['Unknown source' if pd.isna(data.iloc[x+1]['VALUE'])==True and pd.isna(data.iloc[x+2]['VALUE'])==True else 'All sources of diverted materials' for x in unknown_source.index ]

unknown_source = unknown_source[unknown_source['Sources of materials prepared for recycling']=='Unknown source']
unknown_source
```

```python
# append unknown source rows to main df
waste_can = pd.concat([data, unknown_source])
waste_can.tail()
```

```python
# calculate distribution between sources for all materials diverted
pivot_df = waste_can[waste_can['Type of materials diverted']!="All materials diverted"].groupby(['REF_DATE', 'GEO', 'Sources of materials prepared for recycling']).sum()['VALUE']
pivot_df
```

```python
# paste sum of all materials diverted for residenatial and non-residential sources
for index, row in waste_can[(waste_can["Type of materials diverted"]=="All materials diverted") & (waste_can['Sources of materials prepared for recycling']!='All sources of diverted materials')].iterrows():
    waste_can.loc[index, 'VALUE'] = pivot_df.loc[row['REF_DATE'], row['GEO'], row['Sources of materials prepared for recycling']]
```

```python
# as some values are suppressed for specific materials, but still included in all materials row, 
# All sources of materials diverted > sum(residential, non-residential, unknown source)
# Means the difference should be considered as unknown source
# unknown source for all materials diverted = All sources - residetinal source - non-residential source
waste_can[waste_can["Type of materials diverted"]=="All materials diverted"]
```

```python

# filter all sources amount for all materials
# used to create new rows with unknown source
unknown_source_all_materials = waste_can[(waste_can["Type of materials diverted"]=="All materials diverted") & (waste_can['Sources of materials prepared for recycling']=='All sources of diverted materials')].copy()
unknown_source_all_materials
```

```python
# calculate unknown source value
for index, row in unknown_source_all_materials.iterrows():
    non_residential = pivot_df.loc[row['REF_DATE'], row['GEO'], 'Non-residential sources of diverted materials']
    residential = pivot_df.loc[row['REF_DATE'], row['GEO'], 'Residential sources of diverted materials']
    all_sources = row['VALUE'] 
    unknown_source_all_materials.loc[index, 'VALUE'] = all_sources - residential - non_residential
```

```python
# replace name of source to unknown
unknown_source_all_materials['Sources of materials prepared for recycling'] = 'Unknown source'

# append to main table
waste_can = pd.concat([waste_can, unknown_source_all_materials])

waste_can[-40:-15]
```

```python
# export clean data
waste_can.to_csv('waste_can.csv',index=False)
```
